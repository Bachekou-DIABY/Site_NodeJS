```python
global_nuclear_plants_df = global_plants_df[global_plants_df['primary_fuel'] == 'Nuclear']
```

```python
###  Top 20 countries that appear most in the list
n = 20
top20 = global_nuclear_plants_df['country_long'].value_counts()[:n].index.tolist()
top20.reverse()
top20_df = global_plants_df[global_plants_df['country_long'].isin(top20)]
top20_df
```

```python
top_20_nuclear_countries_df = top20_df[(top20_df.primary_fuel == 'Nuclear')]
top_20_nuclear_countries_df
```

```python
number_of_plants = top_20_nuclear_countries_df['country_long'].value_counts()
```

```python
fig = px.histogram(top_20_nuclear_countries_df, y="country_long",color_discrete_sequence = ['darkred'],
    title="Number of nuclear power plants of the top 20 countries")
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
top_20_nuclear_countries_df.describe()
```

```python
top_20_nuclear_countries_df_sum= top_20_nuclear_countries_df.groupby('country_long')['capacity_mw'].sum().reset_index()
top_20_nuclear_countries_df_sum
```

```python
data = dict(type='choropleth', 
            colorscale='Portland',
            z = top_20_nuclear_countries_df_sum['capacity_mw'],
            locations = top_20_nuclear_countries_df_sum['country_long'],
            locationmode ='country names',
            text = top_20_nuclear_countries_df_sum['country_long'],
            marker = dict(line = dict(color='rgb(255,255,255)',width =1)),
            colorbar = { 'title': 'Capacity'})

            
```

```python
layout = dict(title= 'Electrical generating capacity in the top 20 countries',
             geo = dict(scope='world',
                       showlakes=True,
                       lakecolor = 'rgb(85,173,240)')
             )
```

```python
choromap = go.Figure(data= [data], layout=layout)
```

### Capacity is the maximum level of electric power (electricity) that a power plant can supply at a specific point in time under certain conditions.

```python
fuels_capcity_df = global_plants_df[['country_long','name','capacity_mw','primary_fuel']]
```

```python
number_of_plant_all_fuel = fuels_capcity_df['primary_fuel'].value_counts()
```

```python
fuels_capcity_df_sum = fuels_capcity_df.groupby('primary_fuel').agg({'primary_fuel':'count','capacity_mw':'mean'})
fuels_capcity_df_sum
```

```python
fig = px.histogram(fuels_capcity_df,x="primary_fuel", y="capacity_mw",histfunc='avg',title ='Average capacity of different power plants fuel type')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
