```python
 france = df_nuclear_fr[['country_long','name','longitude','latitude']]
```

```python
france
```

```python
# using geopands long and lat to points
france_geo = gpd.GeoDataFrame(france,geometry = gpd.points_from_xy(france.longitude,france.latitude))
```

```python
france_geo
```

```python
#get built on ddata from geopandas
world_data = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
axis = world_data[world_data.continent == 'Europe'].plot(
    color = 'lightblue',edgecolor = 'black')
france_geo.plot(ax = axis,color ='red')
plt.title ('nuclear in France')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(30,30)
fig.savefig('matplot.png',dpi=200)
plt.show()
```
